```python
#We import the pyplot module, and call it plt
import matplotlib.pyplot as plt

# necessary configuration of pyplot to see the images in grayscale
plt.rcParams["image.cmap"]="gray"

# Jupyter command to automatically display images 
%matplotlib inline

# we also import numpy since we use it to create and manipulate arrays
import numpy as np

import matplotlib as mpl
```

**1. BLACK IMAGE**

```python
# size of the arrays to display
size=(20,30)

# An array of zeros
blackImage = np.zeros(size)

#we visualize the matrix
#It looks like a black image, since all elements (pixels) have intensity 0
plt.imshow(blackImage, vmin=0,vmax=1)

# it is necessary to indicate vmin and vmax so that pyplot knows that the minimum is 0 and the maximum 1
```

**2. WHITE AND GRAY IMAGE**

```python
#BLACK
#An array of ones
whiteImage=np.ones(size)

#We visualize the matrix
#It looks like a white image, since all elements (pixels) have intensity 1
plt.imshow(whiteImage, vmin=0, vmax=1)

# We create another figure to show the image (otherwise the next imshow overwrites the previous one
plt.figure()

#GRAY
# An array with value 0.5 in all its elements
grayImage=np.ones(size)*0.5

#It looks like a gray image, since all elements (pixels) have intensity 0.5
plt.imshow(grayImage, vmin=0, vmax=1)
```

**3. GRAY DARK AND RANDOM IMAGE**

```python
# An array with value 0.2 in all its elements
grayDarkImage=np.ones(size)*0.2 #Looks like a gray dark because all his elements (pixels) have an intensity of 0.2
plt.imshow(grayDarkImage, vmin=0, vmax=1)
plt.figure()

#An array with random value
randomImage=np.random.rand(size[0],size[1])
plt.imshow(randomImage, vmin=0, vmax=1)
```

```python
from skimage import io

image=io.imread("/Users/adrianova/Desktop/Data Science/Python/IT Academy - Python/SPRINT 3 - Estructura de dades/S03 T02 Estructura d'una Matriu/images/lena.png") #imread -> read an image with codified pixels like enter numbers
plt.imshow(image) #no es necesario el vmin y el vmax para imagenes a color

print(" Image dimensions: \n", image.shape)
```

 **EXTRACT CHANNELS**

```python
#We can see each channel individually: R,G,B
plt.imshow(image[:,:,0])
plt.title("RED CHANNEL")
plt.figure()

plt.imshow(image[:,:,1])
plt.title("GREEN CHANNEL")
plt.figure()

plt.imshow(image[:,:,2])
plt.title("BLUE CHANNEL")
```

```python
# Another way to view the information for each channel is to remove the information from the other channels. For example, to see the red channel, we set the blue and green channels to 0.
redImage=np.copy(image) #I create a copy of the image to preserve the original

redImage[:,:,1]=0 #Removing color green 
redImage[:,:,2]=0 #Removing color blue

plt.title("RED CHANNEL")
plt.imshow(redImage)

# We can see that without the blue and green channels we lose a lot of color information, although the outline remains.
```

```python
#The same but only removing the blue channel
redAndGreenImage=np.copy(image)

redAndGreenImage[:,:,2]=0 #Removing color blue

plt.title("WITHOUT BLUE")
plt.imshow(redAndGreenImage)
```
